```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\Iris.csv")
df
```

```python
x=df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
print(df["Species"].unique())
label={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
y=[label[c] for c in df["Species"]]
```

```python
#kmeans model
k=KMeans(n_clusters=3)
k.fit(x)
```

```python
colors=np.array(["blue","yellow","green"])
```

```python
#kmeans plot
plt.scatter(df["PetalLengthCm"],df["PetalWidthCm"],c=colors[k.labels_],s=40)
plt.title("kmeans clustering")
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
#gmm model
from sklearn.mixture import GaussianMixture
k2=GaussianMixture(n_components=3,random_state=0).fit(x)
p=k2.predict(x)
```

```python
plt.scatter(df["PetalLengthCm"],df["PetalWidthCm"],c=colors[p])
plt.title("gmm classification")
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
from sklearn import metrics
#accuarcy_kmeans
print("the accuarcy of kmeans is:",metrics.accuracy_score(y,k.labels_))
#accuracy_gmm
print("the accuarcy of gmm is:",metrics.accuracy_score(y,p))
```
